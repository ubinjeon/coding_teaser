002. 평균 구하기

세준이는 기말고사를 망쳤다. 그래서 점수를 조작해 집에 가져가기로 결심했다. 일단 세준이는 자기 점수 중 최댓값을 골랐다. 그런 다음 최댓값을 M이라 할 때 모든 점수를 점수 / M * 100으로 고쳤다. 예를 들어 세준이의 최고점이 70점, 수학 점수가 50점이라면 수학 점수는 50/70*100이므로 71.43점이다. 세준이의 성적을 이 방법으로 계산했을 때 새로운 평균을 구하는 프로그램을 작성하시오.

```python
3 # 시험을 본 과목의 개수
40 80 60 # 각 과목의 시험 성적
```

```python
N = int(input("시험을 본 과목의 개수: "))
scores = input("각 과목의 시험 성적: ").split(" ")


M = 0
for s in scores:
  if int(s) > M:
    M = int(s)

sum = 0
for i in range(len(scores)):
  new_score = int(scores[i])/M*100
  sum += new_score

mean = sum/len(scores)
print(mean)
```

```python
# 오답노트
# 1. sum 은 함수가 있다! 함수로 자료형 이름 선언하지 말기!
# 2. 변환 점수를 일일이 구할 필요 없이 한 번에 변환 점수의 평균 점수를 구할 수 있다.
# (A/M*100 + B/M*100 + C/M*100)/3 = (A+B+C) * 100 / M / 3
# 3. map 함수의 기본 문법 다시 되집기: map(function, iterable)
n = input()
mylist = list(map(int, input().split()))
mymax = max(mylist)
sum = sum(mylist)
print(sum*100/mymax/int(n))
```

003. 구간 합 구하기 1

수 N개가 주어졌을 때 i번째 수에서 j번째 수까지의 합을 구하는 프로그램을 작성하시오.

```python
5 3 # 데이터의 개수, 질의 개수
5 4 3 2 1 # 구간 합을 구할 대상 배열
1 3
2 4
5 5
```

```python
nm = list(map(int, input().split()))
a_list = list(map(int, input("배열의 수를 차례로 입력하시오: ").split()))

s_list = [0]*nm[0] # 구간합
for i in range(len(s_list)):
  if i == 0:
    s_list[i] = a_list[i]
  else:
    s_list[i] = a_list[i] + s_list[i-1] # 합 배열 S를 만드는 공식: S[i] = S[i-1] + A[i]

for m in range(nm[1]):
  ij = list(map(int,input().split()))
  if ij[0] - 1 == 0:
    ij_sum = s_list[ij[1] - 1]
  else:
    ij_sum = s_list[ij[1] - 1] - s_list[ij[0]- 2] # S[j] - S[i-1] i에서 j까지 구간 합
  print(ij_sum)
```

```python
# 오답 노트
# import sys
# input = sys.stdin.readline
suNo, quizNo = map(int, input().split())
numbers = list(map(int, input().split()))
prefix_sum = [0] # 여기서는 prefix_sum 즉 합배열 인덱스 수가 numers 인덱스 수보다 크게 한 듯. 즉 S_ij 에서 i랑 j 가 0보다 큰 몇번째 자리의 수인 듯
temp = 0

for i in numbers:
  temp = temp+i
  prefix_sum.append(temp) # 합 배열 만들기

for i in range(quizNo):
  s, e = map(int, input().split())
  print(prefix_sum[e] - prefix_sum[s-1])
```

004. 구간 합 구하기 2

N x N 개의 수가 N x N 크기의 표에 채워져 있다. 표 안의 수 중 ($X_1$, $Y_1$)에서 ($X_2$, $Y_2$)까지의 합을 구하려고 한다. 표에 채워져 있는 수와 합을 구하는 연산이 주어졌을 때 이를 처리하는 프로그램을 작성하시오.

```python
n, m = map(int, input().split())
A = [[0]*(n+1)]
D = [[0]*(n+1) for _ in range(n+1)]

for i in range(n):
  A_row = [0] + [int(x) for x in input().split()]
  A.append(A_row)

for i in range(1, n + 1):
  for j in range(1, n + 1):
    # 합 배열 구하기
    D[i][j] = D[i][j-1] + D[i-1][j] - D[i-1][j-1] + A[i][j]

for _ in range(m):
  x1, y1, x2, y2 = map(int, input().split())
  # 구간 합 배열로 질의에 답변
  result = D[x2][y2] - D[x1-1][y2] - D[x2][y1-1] + D[x1-1][y1-1]
  print(result)
```

005. 나머지 합 구하기

N개의 수 A1, A2, ..., AN이 주어졌을 때 연속된 부분의 합이 M으로 나누어떨어지는 구간의 개수를 구하는 프로그램을 작성하시오. 즉, Ai + ... + Aj(i <= j)의 합이 M으로 나누어떨어지는 (i,j) 쌍의 개수를 구하시오.

```python
n, m = map(int, input().split())
A = list(map(int, input().split()))
S = [0] * n
C = [0] * m
S[0] = A[0]
answer = 0

for i in range(1, n):
  S[i] = S[i-1] + A[i] # 합 배열 저장

for i in range(n):
  remainder = S[i] % m # 합 배열의 모든 값에 % 연산 수행
  if remainder == 0: # 0~i까지의 구간 합 자체가 0일 때 정답에 더하기
    answer += 1
  C[remainder] += 1 # 나머지가 같은 인덱스의 개수 값 증가시키기

for i in range(m):
  # 나머지가 같은 인덱스 중 2개를 뽑는 경우의 수를 더하기
  if C[i] > 1:
    answer += (C[i] * (C[i] - 1) // 2)

print(answer)
```
